**1. Tentukan data yang akan anda gunakan dalam visualisasi**

```python
import pandas as pd
data = pd.read_csv('/content/ElectricCarData_Clean.csv')
data.head()
```

**2. Berdasarkan data yang anda pilih tentukan tujuan, user, fungsi dan tone (jika ada) yang sesuai dengan visualisasi yang akan anda buat**

### Tujuan
Menampilkan data mobil listrik berdasarkan berbagai faktor seperti jangkauan, efisiensi, kecepatan, dan harga (dalam euro). Visualisasi akan membantu analisis dalam pemilihan keputusan untuk membeli mobil listrik atau hanya sekedar menambah pengetahuan.
### User
Penggemar teknologi dan otomotif, calon pembeli mobil listrik.

**3. Buatlah visualisasi dengan mengunakan library python seperti matplotlib, seaborn atau tools lain**

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PriceEuro', y='Range_Km', hue='Brand', data=data, palette='tab10', s=100, alpha=0.7)
plt.title('Hubungan antara harga dan jarak tempuh')
plt.xlabel('Price (Euro)')
plt.ylabel('Range (Km)')
plt.legend(loc='upper right', bbox_to_anchor=(1.15, 1), title="Brand", ncol=1)
plt.tight_layout()
plt.show()
```

Terlihat bahwa pada scatter plot, semakin tinggi harga, semakin jauh pula jarak tempuh dari mobil listrik.
